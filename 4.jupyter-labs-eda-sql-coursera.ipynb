```python
import pandas as pd
import time
time.clock = time.time
from pandasql import sqldf
df = pd.read_csv('Spacex.csv')
```

```python
pysqldf = lambda q: sqldf(q, globals())
pysqldf
```

```python
df.head()
```

```python
df.columns = ['Date', 'Time_(UTC)', 'Booster_Version', 'Launch_Site', 'Payload',
       'PAYLOAD_MASS__KG_', 'Orbit', 'Customer', 'Mission_Outcome',
       'Landing_Outcome']
```

## Tasks

Now write and execute SQL queries to solve the assignment tasks.

### Task 1




##### Display the names of the unique launch sites  in the space mission


```python
pysqldf('select distinct Launch_Site from df')
```


### Task 2


#####  Display 5 records where launch sites begin with the string 'CCA' 


```python
pysqldf("select * from df where Launch_Site like 'CCA%' limit 5")
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
pysqldf("select sum(PAYLOAD_MASS__KG_) from df where Customer='NASA (CRS)'")
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
pysqldf("select avg(PAYLOAD_MASS__KG_) from df where Booster_Version like 'F9 v1.1%'")
```

### Task 5

##### List the date when the first successful landing outcome in ground pad was acheived.


_Hint:Use min function_ 


```python
pysqldf("select min(Date) from df where Landing_Outcome='Success'")
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
pysqldf("select * from df where Mission_Outcome='Success' and PAYLOAD_MASS__KG_ between 4000 and 6000")
```

### Task 7




##### List the total number of successful and failure mission outcomes


```python
pysqldf("select Mission_Outcome, count(Mission_Outcome) from df group by Mission_outcome")
```

### Task 8



##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
pysqldf("select distinct Booster_Version from df where PAYLOAD_MASS__KG_ = (select max(PAYLOAD_MASS__KG_) from df)")
```

### Task 9

### Task 10

##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order


```python
pysqldf("select Landing_Outcome,count(Landing_Outcome),rank() over (order by count(*) desc) from df where Landing_Outcome in ('Failure (drone ship)','Success (ground pad)') and date between '04-06-2010' and '20-03-2017' group by Landing_Outcome")
```
